```python
import pandas as pd

# Ruta del archivo CSV original (cambia el nombre si es necesario)
csv_file_path = 'Most Streamed Spotify Songs 2024.csv'  # Cambia este nombre si el archivo tiene otro nombre

spotify_data = pd.read_csv(csv_file_path, encoding='ISO-8859-1')

print("Vista preliminar de los datos:")
display(spotify_data.head())

# Limpiar columnas numéricas y convertirlas correctamente
numeric_columns = ['Spotify Streams', 'YouTube Views', 'TikTok Likes', 'Track Score']

for col in numeric_columns:
    # Verificar si la columna es de tipo string, si no es, la convertimos a string primero
    if spotify_data[col].dtype == 'object':
        spotify_data[col] = spotify_data[col].str.replace(',', '', regex=False)
    
    # Convertir la columna a tipo float
    spotify_data[col] = pd.to_numeric(spotify_data[col], errors='coerce')

# Convertir la columna de fechas a formato de fecha
spotify_data['Release Date'] = pd.to_datetime(spotify_data['Release Date'], errors='coerce')

# Limpiar la columna 'Explicit' y convertirla a valores binarios (1 = Sí, 0 = No)
spotify_data['Explicit Track'] = spotify_data['Explicit Track'].apply(lambda x: 1 if x == 'Yes' else 0)

# Eliminar filas con valores nulos
spotify_data_cleaned = spotify_data

# Mostrar las primeras filas del dataset limpio
print("Datos después de limpiar:")
display(spotify_data_cleaned.head())

# Guardar el dataset limpio en un nuevo archivo CSV
output_file_path = 'spotify_data_cleaned.csv'
spotify_data_cleaned.to_csv(output_file_path, index=False)
```
